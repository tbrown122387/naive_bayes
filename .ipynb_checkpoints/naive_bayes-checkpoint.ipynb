# Getting Started with Naive Bayes 



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

## Exploring the Data

It's easier to visualize the data if there are only two predictors. We will consider both the full data set with 13 predictors (`df`), and a smaller data set with only two predictors (`baby_df`).

```python
# get a data set
X, y = load_wine(return_X_y=True)

# optional: view it 
df = pd.DataFrame(np.concatenate((X,y.reshape((178,1))), axis=1))
df.columns = [ 'x'+str(i+1) for i in range(13)] + ['y'] 
df.sample(10).head()
```

```python
# show a random selection of rows
baby_df = df[['x1','x2','y']]
baby_df.sample(10).head()
```

```python
# make a plot of just these two predictors (and class labels)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='viridis');
```

## checking assumptions

We need to check conditional normality of the predictors, and conditional independence.

```python
# checking independence (naiveness)
df[df.y == 0].drop(['y'], axis=1).corr()
#df[df.y == 1].drop(['y'], axis=1).corr()
#df[df.y == 2].drop(['y'], axis=1).corr()
# more advanced option: hypothesis tests
```

```python
# check normality
df[df.y==0].drop(['y'], axis=1).hist()
#df[df.y==1].drop(['y'], axis=1).hist()
#df[df.y==2].drop(['y'], axis=1).hist()
```

## Estimating the parameters


There are mean and standard deviation parameters for each predictor/label pair.

```python
# all predictors
gnb = GaussianNB()
estimated_model_all_preds = gnb.fit(X,y)
estimated_model_all_preds.class_prior_
# estimated_model_all_preds.theta_ # mean params
# estimated_model_all_preds.sigma_ # std. dev. params
```

## Prediction

Plug in estimated parameters, and new predictor data to predict an unseen label.

```python
#prediction on test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)
y_pred = GaussianNB().fit(X_train, y_train).predict(X_test)
print("Number of mislabeled points out of a total %d points : %d"% (X_test.shape[0], (y_test != y_pred).sum()))
```

```python
# make the same plot as above but add boundaries
# reworked code from 
fig, ax = plt.subplots()

ax.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='viridis')

eps = 1
xlim = (min(X[:, 0])-eps, max(X[:, 0]) + eps)
ylim = (min(X[:, 1])-eps, max(X[:, 1]) + eps)

xg = np.linspace(xlim[0], xlim[1], 60)
yg = np.linspace(ylim[0], ylim[1], 40)
xx, yy = np.meshgrid(xg, yg)
Xgrid = np.vstack([xx.ravel(), yy.ravel()]).T

for label, color in enumerate(['red', 'blue', 'orange']):
    mu = np.array([
        estimated_model_all_preds.theta_[label,0], 
        estimated_model_all_preds.theta_[label,1]
    ])
    std = np.array([
        estimated_model_all_preds.sigma_[label, 0],
        estimated_model_all_preds.sigma_[label, 1]
    ])
    P = np.exp(-0.5 * (Xgrid - mu) ** 2 / std ** 2).prod(1)
    Pm = np.ma.masked_array(P, P < 0.03)
    ax.pcolorfast(xg, yg, Pm.reshape(xx.shape), alpha=0.5,
                  cmap=color.title() + 's')
    ax.contour(xx, yy, P.reshape(xx.shape),
               levels=[0.01, 0.1, 0.5, 0.9],
               colors=color, alpha=0.2)
```
